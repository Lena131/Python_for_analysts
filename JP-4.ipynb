## *Условие 2: 2 задача*

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['no', 'yes']);
```

#### *Вывод: преобладает недвижимость без вида на набережную, что всязано с разницей в стоимости такой недвижимости.*

### 2.2 Изучите распределение этажей домов

```python
df_fl = df['floors'].value_counts()
df_fl
```

```python
sns.barplot(x = df_fl.index, y = df_fl)
plt.title('Распределение этажей домов')
plt.xlabel('Кол-во этажей')
plt.ylabel('Кол-во')
plt.xticks(rotation=45);
```

#### *Вывод: преобладает недвижимость c низкой этажностью: на первом месте - 1-этажные, на втором - 2-этажные дома.*

### 2.2 Изучите распределение состояния домов

```python
df_con = df['condition'].value_counts()
df_con
```

```python
sns.barplot(x = df_con.index, y = df_con)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние дома')
plt.ylabel('Кол-во');
```

## *Условие 3: 3 задача*

### Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
### Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg');
```

#### *Вывод: чем больше жилплощадь дома, тем выше его стоимость*

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg');
```

#### *Вывод: чем выше качество конструкции и дизайна дома, тем выше его стоимость*

```python
sns.boxplot(x=df['price'], y=df['grade'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('grade')
plt.title('Distribution of price by grade');
```

#### *Вывод: наибольшую стоимость имеют дома с оценкой конструкции и дизайна 13*

```python
sns.boxplot(x=df['price'], y=df['floors'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('floors')
plt.title('Distribution of price by floors');
```

#### *Вывод: Наибольшую стоимость имеют дома с количеством этажей 2,5*

```python
sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5);
plt.xlabel('price')
plt.ylabel('view')
plt.title('Distribution of price by view');
```

#### *Вывод: Наибольшую стоимость имеют дома с оценкой вида 4*
